```python
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
```

```python
df
```

```python
df['W']
# THIS RETURNS A SERIES. HENCE A DATAFRAE IS A SET OF SERIES DISPLAYED IN TABULATURE FORMAT.
```

```python
df[['W','Z']]
# WHEN YOU CALL FOR TWO COLUMNS, IT RETURNS A MINI/ SUB-DATAFRAME...
```

```python
# TO ADD A COLUMN TO THE DATAFRAME, WE USE THE NORMAL WAY OF CALLING A COLUMN.
# WE ALSO USE ARITHMETIC FOR EASIER CONSTRUCTION OF  COLUMN AS SHOWN BELOW..........
df['R'] = df['W']+df['X']
```

```python
df
```

```python
# TO DROP A COLUMN FROM THE TABLE, WE USE A SIMILAR SYNTAX BUT THIS TIME SLIGHTLY DIFFERENT.
# WE USE THE 'drop' FUNCTION IN PANDA AND SPECIFY THE AXIS........
df.drop('Y', axis=1)
```

```python
# IT DISPALYS THE ORIGINAL DATAFRAME WITHOUT THE Y COLUMN BUT IN MEMORY THE Y COLUMN STILL EXISTS.....
df
```

```python
# TO REMOVE IT WE USE THE 'inplace' PARAMETER INLINE WITH THE 'drop' PANDAS FUNCTION........
df.drop('Y', axis=1, inplace=True)
```

```python
# THE PARAMETER HAS TO BE SET TO TRUE IF YOU ARE TO MAKE PERMANENT ALTERATIONS IN PYTHON.
df
```

```python
df.loc['B']
```

```python
# THE SYNTAX ABOVE SELECTS ROWS AND RETURNS THEM AS SERIES WITH THE COLUMN LETTERS BEING THE INDEX LABELS
# IF ONE IS NOT SURE IF THE ROWS LABELS ARE STRINGS THEN THEY CAN USE INDEX REFERENCING.
df.iloc[1]
```

```python
# TO LOCATE A SPECIFIC VALUE IN A CELL. WE USE COMMA NOTATION AND SPECIFY THE ROWS AND COLUMNS....
df.loc['C','R']
```

```python
# YOU CAN ALSO SELECT DIFFERENT VALUES IN DIFFERENT COLUMNS AND ROWS RETURNING THEM AS A DATAFRAME
df.loc[['A','C','D'],['W','R']]
```

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside, inside))
```

```python
hier_index
```

```python
hier_index1 = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index1
```

```python
df2 = pd.DataFrame(randn(6,2), hier_index1, ['A','B'])
```

```python
df2
```

```python
''' 
HIER_INDEX HERE IS MADE UP. BUT IT STANDS FOR HIGHER INDEX or HIERACHY.
YOU NOTICE IN THE PREVIOUS SCRIPT WE HAVE MULTILEVEL OF THE INDEX WITH OUR FIRST LEVEL INDICES BEING G1 AND G2
THE NEXT LEVEL IS 1, 2, 3 FOR EACH OF THE FIRST LEVEL INDEX LABELS.
JUST AS SEEN IN LINE 57.
''' 
```

```python
df2.loc['G1']
```

```python
df2.loc['G2'].loc[2]
# THIS SCRIPT RETURNS A SERIES OF THE ROW 2 IN INDEX LABEL G2.........
```

```python
# TO LABEL THE INDEX LEVELS, WE USE THE SCRIPT BELOW..............
df2.index.names = ['Groups','Num']
```

```python
df2
```

```python
# TO CALL A MULTILEVEL INDEX DATAFRAME WE USE THE SCRIPT BELOW..........
df2.loc['G2'].loc[3]['A']
```

```python
# TO DO CROSS-SECTIONAL CELL SELECTION OR CALLING, WE USE THE SCRIPT BELOW.
# TAKE AN EXAMPLE OF SELECTING BOTH ROWS OF '1' IN BOTH G1 AND G2.........
df2.xs(1, level='Num')
```
