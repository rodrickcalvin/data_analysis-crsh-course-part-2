```python
import pandas as pd
```

```python
df = pd.DataFrame({'col1':[1,2,3,4],
                   'col2':[444,555,666 ,444],
                   'col3':['abc','def','ghi','xyz']})
df.head()
```

```python
df['col2'].unique()
```

```python
# TO SHOW THE NUMBER OF UNIQUE VALUES. WE USE THE SYNTAX BELOW
df['col2'].nunique()
```

```python
# SERIES OF UNIQUE VALUES AND HOW HMUCH THEY OCCUR
df['col2'].value_counts()
```

```python
# PANDAS NOT ONLY SUPPPORTS BUILT-IN FUNCTIONS BUT ALSO USER-DEFINED FUNCTIONS.
# TO SAVE TIME WE INSTEAD USE A LAMBDA EXPRESSION INSTEAD OF A FUNCTION.
# TO USE THE FUNCTION, WE USE THE PANDAS 'apply' METHOD.
df['col2'].apply(lambda x: x*2)
```

```python
# TO ORDER YOUR DATAFRAME BY A SPECIFIC COLUMN.
df.sort_values(by='col2')
```

```python
# TO FIND OUT ALL THE NULL VALUES IN A DATAFRAME WE USE
df.isnull()
```
