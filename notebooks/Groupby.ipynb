```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'], 'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'], 'Sales':[200, 120, 340, 124, 243, 350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
query1 = df.groupby('Company')
```

```python
query1.mean()
```

```python
'''
Group by helped us focus on the company column. It then carries out manipulations like average of the fields
that are attached to specific companies considering they are integers.
'''
```

```python
query1.std()
```

```python
# we can also carry out the standard deviation of every unique company in reference to their name in the
# DataFrame we created.
```

```python
query1.sum()
```

```python
query1.sum().loc['FB']
```

```python
# TO GET THE SUM OF THE SALES IN FB WE USE THE SCRIPT ABOVE OR THE ONE LINE SCRIPT BELOW....
df.groupby('Company').sum().loc['FB']
```

```python
# TO COUNT THE NUMBER OF UNIQUE INSTANCES IN THE COLUMNS WITH RESPECT TO THE UNIQUE COMPANY NAMES WE USE:
df.groupby('Company').count()
```

```python
# TO GET THE MIN AND MAX
df.groupby('Company').max()
df.groupby('Company').min()
```

```python
# TO GET ALL THE INFORMATION AT ONCE I CAN USE THE FOLLOWING SCRIPT:
df.groupby('Company').describe()
```
